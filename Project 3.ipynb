# 16S & 23S rRNA Comparison of M. Pneumoniae and Gut-Dwelling Archaea Species

### Scientific Question: Is there a similarity in 16s, 23s rRNAs of pathogenic Mycoplasmas pneumoniae and Gut-Dwelling Archaea Species Candidatus Methanomassiliicoccus Luminyensis, Candidatus Methanomassiliicoccus intestinalis and Candidatus Methanomethylophilus alvus?

With the cost of DNA sequencing decreasing by the passing year, gene comparisons have become a prevailing factor in identifying¬¬ bacteria and archaea species. Some regions of the genome are more highly conserved than others, such regions for bacteria and archaea domains are 16s / 23s ribosomal RNA (Clarridge et al., 2014). 16s / 23s rRNA are only present in bacteria and archaea domains and are highly conservative with low but constant mutation rate. These characteristics allow the region to be used for species identification in bacteria and archaea domains. A local pairwise alignment was used to compare a species of pathogenic bacteria and a group of gut-dwelling of archaea that have similar characteristics.

Mycoplasma pneumoniae is a pathogenic bacteria species that causes pneumonia and chronic fatigue syndrome (Jiang et al., 2021). A unique characteristic of M. pneumonia is that this species of bacteria evolved to have no peptidoglycan cell wall (Hu, P C et al., 1975). Most bacteria species have a thick (gram positive) or thin (gram negative) peptidoglycan cell wall as the outermost defense mechanism. The fact that most bacteria have a layer of peptidoglycan became a popular target for antibacterial research. However, the lack of gram-cell wall is an evolutionary advantage for M. pneumonia, as the bacteria developed adhesins, organelle that attach to the epithelial layer, for effective infection of the host eukaryotic cell. The lack of peptidoglycan protects the species from most antimicrobials as well, which makes M. pneumonia developed pneumonia or related infections very difficult to treat once the bacterial colony grows.

Although classified as a bacteria species, M. pneumoniae’s lack of gram cell wall more closely resembles archaea species, as the domain’s key characteristic is the S-layer only cell wall structure. Some archaea are closer to the structure of M. pneumonia in that they also lack the cell wall structure. Such species are the thermoplasmatales which dwell in very extreme environments such as high / low pH or extreme temperatures. One related genus of these archaea live in the human guts: Methanomassiliicoccales. Three were identified, C. M. luminyensis, C. M. intestinalis, and C. M. alvus. C. M. lyminyensis was able to thrive in soil environment as well, while the others were very specific to gut biome (Borrel, Guillaume et al., 2014).

A previous research compared the spike protein sequence and the gene encoding for the spike proteins of various betacoronavirus strains from different animal sources (Javier A. Jaimes et al., 2020). The similarity in the sequences generated a phylogenetic tree that revealed pangolin and bat strains of coronavirus was similar to the human SARS-CoV-2 strain. Similarly, this project aims to compare the 16s and 23s rRNA gene of M. pneumoniae to 16s / 23s rRNA gene of C. M. luminyensis, C. M. intestinalis, and C. M. alvus. 

Three M. pneumoniae 16s rRNA and one 16s rRNA per archaea species are provided for the first local pairwise alignment comparison. Three M. pneumoniae 23 rRNA and one 23 rRNA per archaea species are provided for the second local pairwise alignment comparison. All 16s / 23s rRNA gene sequences are from GenBank, the NIH genetic sequence database. GenBank is part of “International Nucleotide Sequence Database Collaboration” with the DataBank of Japan and European Nucleotide Archive. As of February 2021, there are 226 million sequences in GenBank alone, and 1.5 billion sequences in Whole Genome Sequencing (WGS) division. Genes of interest can be searched by keywords at GenBank website (https://www.ncbi.nlm.nih.gov/genbank/). 

### Hypothesis: A relatively high similarity in 16s and 23s rRNA comparisons of M. pneumoniae and C. M. luminyensis, C. M. intestinalis and C. M. alvus suggest these two groups had a common evolutionary ancestor that lacks cell walls and dwelled in the human body.


Genbank provided for three partial 16s rRNA gene sequence, and one 23s rRNA gene sequence of M. pneumoniae, as well as one 16s rRNA and one 23s rRNA sequences for each archaea species. M. pneumoniae sequences were found by manually searching the keywords “Mycoplasma pneumoniae” with corresponding rRNA gene (16s/23s) and selecting to download sequences as FASTA files. Archaea 16s / 23s sequences were found by searching the whole genome and selectively looking for 16s regions, and downloading the selected region via “selected region” function on GenBank search. These FASTA files were loaded as strings using SeqIO package from Biopython.

Using pairwise2 package from Biopython, local alignments will be performed between each of the given 16s / 23s rRNA sequences. The alignment scores are normalized by the length of the longer sequence to give a rough comparative percentage similarity between the rRNA gene sequences. Heat maps of the rRNA gene comparisons will be implemented to visually show the similarity percentages. Instructions on implementing the heat maps were provided by matplotlib documentations (https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html). Dot plots of pairwise alignments pair that revealed high similarity, high defined by higher percentage than lowest positive control (explained below), are provided in appendix for further examination of the similarity in genes. Instructions for dot plots are provided in Biopython cookbook provided with the documentation of the package (http://biopython.org/DIST/docs/tutorial/Tutorial.html#sec384). This approach outlined by Biopython allows for easier view of “dots” by using dictionaries instead of NumPy array to draw the dot plot.

Sequencing loop works that each rRNA sequences are put into a 1-D tuple and are referenced for each sequencing to be put into a 2-D NumPy array displaying the alignment score. Each rRNA sequence is thus aligned with itself, as well as the sequences of same domain group. Alignment within the domain group species (such as CMI vs. CMA) acts as a positive control in this experiment. The rRNA similarity within the archaea or bacteria domain will act as a standard to judge whether the alignment similarity across archaea and bacteria is due to the difference in domains or evolutionary similarity.

## Part 3.1: Perform local pairwise alignment of 16s rRNAs of *M. Pneumoniae* and archaea species.

In the section below, parsed nucleotide sequences are organized, then "local pairwise align"-ed. Results are stored into a NumPy array as part of the loop.

This is directly taken from above hypothesis section.
Using pairwise2 module from Biopython, local alignments will be performed between each of the given 16s / 23s rRNA sequences. The alignment scores are normalized by the length of the longer sequence to give a rough comparative percentage similarity between the rRNA gene sequences.
Sequencing loop works that each rRNA sequences are put into a 1-D tuple and are referenced for each sequencing to be put into a 2-D NumPy array displaying the alignment score. Each rRNA sequence is thus aligned with itself, as well as the sequences of same domain group. Alignment within the domain group species (such as CMI vs. CMA) acts as a positive control in this experiment. The rRNA similarity within the archaea or bacteria domain will act as a standard to judge whether the alignment similarity across archaea and bacteria is due to the difference in domains or evolutionary similarity.

```python
#creating tuple of 16s records
record_group_16s = (s16MP1,s16MP3,s16MP4,s16CMA1,s16CMI1,s16CML1)

#matrix of alignments in numpy array
matrix_alignments_16s = np.empty((len(record_group_16s),len(record_group_16s)))

                                 
                                 
#for loop to sequence and store into array
for i in range(len(record_group_16s)):
    for j in range(len(record_group_16s)):
        alignments = pairwise2.align.localxx(record_group_16s[i], record_group_16s[j], score_only = True)
        #check if alignment worked
        if str(alignments) == "":
            print("ERRROR in"+str(i)+","+str(j))
            break
        matrix_alignments_16s[i,j] = alignments

print ("Local Pairwise Alignment Score of 16s rRNA")
print (matrix_alignments_16s)

normal = 0
normalized_alignments = matrix_alignments_16s

#normalize the scores
for i in range(len(record_group_16s)):
    for j in range(len(record_group_16s)):
        first = (len(record_group_16s[i]))
        second = (len(record_group_16s[j]))
        if first >= second:
            normal = first
        else:
            normal = second
        
        normalized_alignments[i,j] = normalized_alignments[i,j] / normal
print("")
print("Normalized Alignment Scores into Percentages")
print(normalized_alignments)

#truncate the values for easier viewing
truncated_16s = np.empty((len(record_group_16s),len(record_group_16s)))
truncated_16s = normalized_alignments.round(decimals = 4)

print("")
print("Truncated Normalized Alignment Scores into Percentages")
print (truncated_16s)
```

Below is the implementation of the heatmap function.

```python
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    """
    Create a heatmap from a numpy array and two lists of labels.

    Parameters
    ----------
    data
        A 2D numpy array of shape (N, M).
    row_labels
        A list or array of length N with the labels for the rows.
    col_labels
        A list or array of length M with the labels for the columns.
    ax
        A `matplotlib.axes.Axes` instance to which the heatmap is plotted.  If
        not provided, use current axes or create a new one.  Optional.
    cbar_kw
        A dictionary with arguments to `matplotlib.Figure.colorbar`.  Optional.
    cbarlabel
        The label for the colorbar.  Optional.
    **kwargs
        All other arguments are forwarded to `imshow`.
    """

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")

    # We want to show all ticks...
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    # ... and label them with the respective list entries.
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right",
             rotation_mode="anchor")

    # Turn spines off and create white grid.
    for edge, spine in ax.spines.items():
        spine.set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, **textkw):
    """
    A function to annotate a heatmap.

    Parameters
    ----------
    im
        The AxesImage to be labeled.
    data
        Data used to annotate.  If None, the image's data is used.  Optional.
    valfmt
        The format of the annotations inside the heatmap.  This should either
        use the string format method, e.g. "$ {x:.2f}", or be a
        `matplotlib.ticker.Formatter`.  Optional.
    textcolors
        A pair of colors.  The first is used for values below a threshold,
        the second for those above.  Optional.
    threshold
        Value in data units according to which the colors from textcolors are
        applied.  If None (the default) uses the middle of the colormap as
        separation.  Optional.
    **kwargs
        All other arguments are forwarded to each call to `text` used to create
        the text labels.
    """

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[int(im.norm(data[i, j]) > threshold)])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts
```

This section below plots the heat map of 16s rRNA.

```python
fig, ax = plt.subplots()

im, cbar = heatmap(truncated_16s*100, axis_names_16s, axis_names_16s, ax=ax,
                   cmap="YlGn", cbarlabel="% similarity")
texts = annotate_heatmap(im)

ax.set_title("16s rRNA Comparisons of M. Pneumoniae vs. Gut Archaea Species")

fig.tight_layout()
plt.show()
```

```python
def dotplot(seq1, seq2):
    pylab.cla() 
    window = 7
    dict_one = {}
    dict_two = {}
    for (seq, section_dict) in [ (seq1, dict_one), (seq2, dict_two)]:
        for i in range(len(seq) - window):
            section = seq[i : i + window]
            try:
                section_dict[section].append(i)
            except KeyError:
                section_dict[section] = [i]
    matches = set(dict_one).intersection(dict_two)
    print("%i unique matches" % len(matches))
    x = []
    y = []
    for section in matches:
        for i in dict_one[section]:
            for j in dict_two[section]:
                x.append(i)
                y.append(j)

    #pylab.cla()  # clear any prior graph
    pylab.gray()
    pylab.scatter(x, y)
    pylab.xlim(0, len(s16MP1) - window)
    pylab.ylim(0, len(s16CMI1) - window)

    #labeling
    pylab.xlabel("%s (length %i bp)" % (print_var_name(seq1), len(seq1)))
    pylab.ylabel("%s (length %i bp)" % (print_var_name(seq2), len(seq2)))

    pylab.title("Dot plot of " + print_var_name(seq1)+ " and " + print_var_name(seq2) + " using window size %i\n(allowing no mis-matches))" % window)
    pylab.show()
```
